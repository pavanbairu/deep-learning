```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
```

```python
X,y = make_moons(n_samples=300, noise=0.1, random_state=42)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
model = Sequential()
```

```python
model.add(Dense(10, activation='sigmoid', input_dim=2))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.get_weights()
```

```python
# 1st layer weights
old_weights = model.get_weights()[0]
old_weights
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.25, random_state=42)
```

```python
model.fit(X_train, y_train, epochs=100)
```

```python
model.get_weights()
```

```python
# taking the 1st layer weights after uppdate
new_weights = model.get_weights()[0]
new_weights
```

```python
old_weights, new_weights
```

#### there is not much changes in the weight updates before and after

```python
model = Sequential()

## adding only few hidden layers
model.add(Dense(10, activation='sigmoid', input_dim=2))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])

old_weights_1 = model.get_weights()[0]

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.25, random_state=42)

model.fit(X_train, y_train, epochs=100)

new_weights_1 = model.get_weights()[0]

old_weights_1, new_weights_1
```

#### here we can observe the loss is 0.4308 earlier it was 0.69 and also quite big changes in the weights compared to the previous neural network

## Using Relu Activation Function

```python
model = Sequential()

model.add(Dense(10, activation='relu', input_dim=2))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.get_weights()
```

```python
o_weights = model.get_weights()[0]
o_weights
```

```python
model.fit(X_train, y_train, epochs=100)
```

#### the loss itself is started from 0.69 and ended with 1.97e-04 using relu activation function. when using sigmoid activation function the least loss was 0.69.
